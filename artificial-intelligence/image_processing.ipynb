```python
import cv2
import seaborn
from matplotlib import pyplot
```

```python
image = cv2.imread("flower.png")
```

```python
pyplot.imshow(image)
pyplot.axis("off")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
red = image.copy()[:, :, 0]
green = image.copy()[:, :, 1]
blue = image.copy()[:, :, 2]

pyplot.imshow(red, cmap='Reds_r')
pyplot.imshow(green, cmap='Greens_r')
pyplot.imshow(blue, cmap='Blues_r')

seaborn.distplot(image.flatten())  #imprime o histograma usando o seaborn
pyplot.xlim([0, 256])
```
