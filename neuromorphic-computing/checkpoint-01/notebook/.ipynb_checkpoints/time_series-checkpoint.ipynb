```python
import requests
import pandas
import numpy

from statsmodels.tsa.arima.model import ARIMA

from matplotlib import pyplot
```

```python
channel_id = 1293177
field_id = 1

response = requests.get(f"https://api.thingspeak.com/channels/{channel_id}/fields/{field_id}.json")
data = response.json()

data
```

```python
dataset = pandas.DataFrame(data["feeds"])

dataset['created_at'] = pandas.to_datetime(dataset['created_at'])
dataset.set_index("created_at", inplace=True)

dataset.drop('entry_id', axis=1, inplace=True)

dataset.rename(columns={"field1": "value"}, inplace=True)

dataset.sample(10)
```
